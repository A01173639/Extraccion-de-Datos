**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto 
#En este filtro se busca que solo aparezcan aquellos registros del vendedor Diego Armando de la columna NOMBRE_VENDEDOR
filtro1=df[df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN"]
filtro1
```

```python
filtro1.to_csv("Filtro1.csv")
```

```python
#Filtro por filas
filtro2= df.iloc[2:19,: ]           #Filas consecutivas
#filtro2=df.iloc[ [8, 24, 100] , : ]  #Filas no consecutivas 
filtro2
#En este filtro guarda solamente las filas de la 2 a la 19
```

```python
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , 1:4]       #Columnas consecutivas 
filtro3
##Filtra la misma cantidad de filas del df original pero limita a las primeras 4 columnas
```

```python
filtro3.to_csv("Filtro3.csv")
```

```python
df1= pd.read_excel('Detalles.xlsx', index_col=1)
df1
#leemos el excel y asignamos la primera columna del df a que sea el indice
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
filtro4=df1.loc[["2022-01-05","2022-12-23"], ["PRECIO_UNITARIO"]]
filtro4
#En este filtro se buscan solo los registros de la columna indice que correspondan a las fechas indicadas,y muestra solo la columna de precio_unitario
```

```python
filtro4.to_csv("Filtro4.csv")
```

```python
#Filtro por cabecera
filtro5=df.head(10)
filtro5
#Imprime los primeros 10 registros del df
```

```python
filtro5.to_csv("Filtro5.csv")
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["CANT"] > 3]
filtro6
#Filtra en la columna "CANT" y muestra solo los valores numericos mayores a 3
```

```python
filtro6.to_csv("Filtro6.csv")
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["PRECIO_UNITARIO"] > 500 ) & (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro7
#Filtra en la columna precio_unitario los valores mayores a 500, que haya hecho la vendedora Leticia (nombre_vendedor)
```

```python
#Filtro o
filtro8=df[(df["PRECIO_UNITARIO"] > 1000 ) | (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
#Filtra en la columna precio_unitario los valores mayores a 500, o en su defecto los que haya hecho el vendedor Alfredo Canela (nombre_vendedor)
filtro8
```

```python
#Filtro not
filtro9=df[(df["PRECIO_UNITARIO"] > 1000 ) & ~(df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]
#En este filtro se busca los registros mayores a mil que no sean del vendedor Diego
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
```
